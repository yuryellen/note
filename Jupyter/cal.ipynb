```python
from sympy import symbols, Function, dsolve, Eq, Derivative

t = symbols('t')
y = Function('y')(t)

# è§£å¸¸å¾®åˆ†æ–¹ç¨‹ï¼šy'' + y' + y = 0
ode = Eq(Derivative(y, t, t) + Derivative(y, t) + y, 0)
solution = dsolve(ode)
print(solution)
```

```python
import numpy as np

# å¾…æ±‚å€¼ç‚¹ x = 1/3
x_val = 1/3

# --- 1. å®šä¹‰å‡½æ•° (ä¸ä¹‹å‰ç›¸åŒ) ---
def func_exp(x):
    """3^x"""
    return np.power(3, x)

def func_neg(x):
    """-x"""
    return -x

funcs = [func_exp, func_neg]

# --- 2. å…³é”®ä¿®æ­£ï¼šå°†æ ‡é‡è½¬æ¢ä¸ºæ•°ç»„ ---
# ä½¿ç”¨ np.array([x_val]) å°†æ ‡é‡è½¬æ¢ä¸º 1 ç»´æ•°ç»„ï¼Œä¿®å¤ IndexErrorã€‚
X_scalar = np.array([x_val])

# é‡æ–°å®šä¹‰æ¡ä»¶ (å¿…é¡»åŸºäº X_scalar)
conds_scalar = [
    X_scalar < 1,           # æ¡ä»¶ 1: x < 1
    X_scalar > 1            # æ¡ä»¶ 2: x > 1
]

# è®¡ç®— f(1/3)
# np.piecewise ä¼šè¿”å›ä¸€ä¸ªåŒ…å«ä¸€ä¸ªå…ƒç´ çš„æ•°ç»„
f_1_3_array = np.piecewise(X_scalar, conds_scalar, funcs)
f_1_3 = f_1_3_array[0] # å–å‡ºæ•°ç»„ä¸­çš„å”¯ä¸€å€¼

# --- 3. ç»“æœéªŒè¯ ---
print(f"ä¿®æ­£åçš„ f(1/3) è®¡ç®—ç»“æœï¼š{f_1_3}")
print(f"ç»“æœä¸º $3^{{1/3}}$ (å³ $\\sqrt[3]{{3}}$)")
print(f"æµ®ç‚¹è¿‘ä¼¼å€¼: {f_1_3:.6f}")
```

```python
from sympy import symbols, Piecewise, Abs

# å®šä¹‰ç¬¦å· x
x = symbols('x')

# å®šä¹‰ç¬¦å·åˆ†æ®µå‡½æ•° f(x)
# æ³¨æ„ Piecewise çš„æ¡ä»¶æ˜¯ (è¡¨è¾¾å¼, æ¡ä»¶)ï¼Œæœ€åä¸€ä¸ªæ˜¯é»˜è®¤å€¼ (True)
f_sympy = Piecewise(
    (3**x, x < 1),    # x < 1 æ—¶ 3^x
    (-x, x > 1),      # x > 1 æ—¶ -x
    (0, True)         # é»˜è®¤å€¼ï¼ˆä¾‹å¦‚ x=1 å¤„ï¼‰
)

# è®¡ç®— f(1/3) çš„ç²¾ç¡®å€¼
result_sympy = f_sympy.subs(x, 1/3)

print("\n## ğŸ”¢ ä½¿ç”¨ SymPy åº“è¿›è¡Œç²¾ç¡®ç¬¦å·è®¡ç®—")
print(f"* f(x) = {f_sympy}")
print(f"* **f(1/3) çš„ç²¾ç¡®ç¬¦å·ç»“æœ: {result_sympy}**")
```

```python
import numpy as np

# å¾…æ±‚å€¼ç‚¹ x = 1/3
x_val = 1/3

# --- 1. å®šä¹‰å‡½æ•° (ä¸ä¹‹å‰ç›¸åŒ) ---
def func_exp(x):
    """3^x"""
    return np.power(3, x)

def func_neg(x):
    """-x"""
    return -x

funcs = [func_exp, func_neg]
# ... å‡½æ•°å®šä¹‰å’Œ funcs åˆ—è¡¨ä¸å˜ ...

# 1. å•ç‚¹è¾“å…¥ (ç”¨äºè®¡ç®— f(1/3) å’Œä¿®å¤ IndexError)
x_val = 1/3
X_scalar = np.array([x_val])
# è®¡ç®— f(1/3)
f_1_3 = np.piecewise(X_scalar, [X_scalar < 1, X_scalar > 1], funcs)[0]

# 2. å¤šç‚¹è¾“å…¥ (ç”¨äºéªŒè¯å’Œæ¼”ç¤ºçŸ¢é‡åŒ–)
X_test = np.array([-2, 0, x_val, 1, 2])
# è®¡ç®— f(X_test)
Y_test = np.piecewise(X_test, [X_test < 1, X_test > 1], funcs)

print(f"f(1/3) çš„ç»“æœ: {f_1_3}")
print(f"X æ•°ç»„ä¸Šçš„æ‰¹é‡è®¡ç®—ç»“æœ: {Y_test}")
```
